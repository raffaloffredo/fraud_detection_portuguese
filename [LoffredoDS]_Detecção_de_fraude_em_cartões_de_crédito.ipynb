# Obteno dos Dados

Os dados utilizados nesse projeto foram disponibilizados por operadoras europeias de carto de crdito. Esse conjunto de dados contm transaes referentes  um perodo de 2 dias, que resultaram em quase 290 mil transaes, nas quais 492 foram classificadas como fraudes. Ou seja, trata-se de um conjunto de dados desbalanceado uma vez que as fraudes representam cerca de 0,17% do total de transaes.
<br/>
<br/>

Na [pgina original dos dados](https://www.kaggle.com/datasets/mlg-ulb/creditcardfraud)  informado ainda que todos os atributos foram convertidos para dados numricos por questes de proteo de dados e descaracterizao dos clientes. . Mesmo o nome das variveis foi modificado e  identificado apenas pela sequncia da letra 'V' seguida do nmero 1, 2, 3, ..., 28. Ou seja, o que se v  apenas V1, V2, V3, ..., V28.
<br/>
<br/>

Adiciona-se ainda, o fato de que essas variveis passaram pelo processo de transformao com o algoritmo no-supervisionado denominado PCA - *Principal Component Analysis* - que tem por objetivo reduzir a dimensionalidade. Isto , um processo que visa diminuir a quantidade original de atributos no conjunto de dados com a menor perda de informaes possvel. Alm disso, o PCA consegue distinguir as informaes relevantes daqueles aleatrias e, at mesmo, redundantes. Por tomar conhecimento desse fato, conclui-se que os dados esto padronizados (esto todos em uma mesma escala), uma vez que isso  pr-requisito para se aplicar o PCA.
<br/>
<br/>

Por fim, o site ressalta que as variveis `Time` e `Amount` no passaram pelo processo de transformao mencionado acima. O atributo `Time` refere-se ao tempo decorrido entre a cada transao e a primeira transao do conjunto de dados. J a caracterstica `Amount`  o valor monetrio de cada transao. Temos tambm a classe desse conjunto, isso , o nosso alvo, que responde a pergunta do nosso problema: "Essa transao  uma fraude?". Para essa varivel a resposta de "Sim" ou "No" est convertida em um valor binrio de 0 quando no se trata de fraude e 1 para quando for reconhecida como fraude.

Primeiro, vou verificar quanto tempo durou o recolhimento dos dados.

```python
# verificar tempo de durao da coleta de dados
print('Durao total da coleta dos dados: {:.0f} segundos'.format(df.Time.max()))
print('Isso corresponde a {:.0f} horas.'.format(df.Time.max() / 3600))
```

Portanto, temos 48h de coleta de dados. Poderemos checar isso hora-a-hora.

```python
# plotar histograma de transaes legtimas x fraudulentas ao longo do tempo

## definir varivel com transaes legtimas x fraudes
legitimate = df[df.Class == 0].Time.div(3600)
fraud = df[df.Class == 1].Time.div(3600)

## configurar histograma
fig, ax = plt.subplots(nrows=2, ncols=1, figsize=(10,4))

## ttulo
ax[0].text(-1.5, 14000, 'Frequncia de Transaes', fontsize=20, color='#004a8f',
        fontweight='bold')

## subttulo
ax[0].text(-1.5, 12500, 'Transaes realizadas com carto de crdito por hora',
           fontsize=9, color='#6d6e70')

## dados
### grfico 1 - transaes legtimas
ax[0].hist(legitimate, label='Legtimas', bins=48, color='#bdbdbd')
ax[0].legend(loc='upper left', frameon=False)
ax[0].spines['right'].set_visible(False)
ax[0].spines['top'].set_visible(False)
ax[0].spines['bottom'].set_visible(False)
ax[0].spines['left'].set_visible(False)
ax[0].set_yticklabels(labels=[], visible=False)
ax[0].set_yticks(ticks=[])

### grfico 2 - transaces fraudulentas
ax[1].hist(fraud, label='Fraudes', bins=48, color='#004a8f')
ax[1].legend(loc='upper left', frameon=False)
ax[1].spines['right'].set_visible(False)
ax[1].spines['top'].set_visible(False)
ax[1].spines['bottom'].set_visible(False)
ax[1].spines['left'].set_visible(False)
ax[1].set_yticklabels(labels=[], visible=False)
ax[1].set_yticks(ticks=[])

plt.show()
```

Observa-se que as transaes legtimas ocorrem com mais fluidez e em determinados horrios. Como no sabemos a hora exata que se iniciaram a coleta dos dados, apenas podemos levantar uma hiptese de que seriam horrios comerciais, onde a populao de forma geral est mais ativa.


Por outro lado, no histograma com a frequncia de fraudes, temos o oposto: no h uma fluidez pois o grfico tem muitos picos. Tambm no h conglomerados de horrios em que se tem mais transaes fraudulentas e, tampouco horrios com menos dessas transaes. Logo, por no apresentar um padro, fica mais difcil reconhec-las.

# Tratamento dos Dados

E antes de realizar alteraes no conjunto de dados, vou copi-lo, para que as alteraes feita daqui por diante sejam realizadas nessa rplica. Assim, o conjunto original  mantido intacto.

### Padronizar Variveis

A padronizao dos dados tem como objetivo deixar todos os valores em uma mesma escala para que atributos diferentes possam ser comparados entre si. Quando os dados no esto padronizados algumas variveis podem causar mais impacto do que outras, prejudicando a performance do modelo j que ele ir dar vantagem a uma varivel em prol de outras. Logo, a padronizao evita que o modelo fique enviesado.

Como visto, as variveis `V1, V2, V3, ..., V28` j passaram pelo PCA e, portanto, infere-se que foram padronizadas, uma vez que esse processo  requisito para a aplicao do PCA nos dados. Portanto, restam as variveis `Time` e `Amount` que no esto padronizadas.

Essa etapa, em condies normais, seria feita em dois momentos diferentes, j com os dados divididos em treino e teste (nos dados de treino, antes de treinar o modelo e, nos dados de teste, antes de serem feitas as previses). Contudo, como o conjunto de dados j passou por um tratamento semelhante em outros dados, portanto, realizar essa etapa antes, no ir trazer maiores prejuzos.

Dessa forma, optou-se por utilizar o ***StandardScaler*** uma vez que ele segue a distribuio normal e transforma a mdia em 0 e o desvio padro em 1.

```python
# padronizar atributos 'Time' e 'Amount' com StandarScaler
## instanciar StandardScaler
std_scaler = StandardScaler()

# criar nova coluna com os dados padronizados em 'Time'
df_clean['Time_STD'] = std_scaler.fit_transform(df_clean['Time'].values.reshape(-1, 1))

# criar nova coluna com os dados padronizados em 'Amount'
df_clean['Amount_STD'] = std_scaler.fit_transform(df_clean['Amount'].values.reshape(-1, 1))

# excluir colunas 'Time' e 'Amount' sem padronizao
df_clean.drop(['Time', 'Amount'], axis=1, inplace=True)

# verificar primeiras entradas do data set modificado
df_clean.head()
```

Podemos checar o resultado da padronizao nos atributos `Time_STD` e `Amount_STD`, com o resumo estatstico:

```python
# imprimir resumo estatstico para 'Time_STD' e 'Amount_STD'
df_clean[['Time_STD', 'Amount_STD']].describe().round(3)
```

Em ambos veja que o Standard Scaler cumpriu sua funo e, agora, temos uma mdia igual a 0 e um desvio padro igual a 1.

### Diviso dos Dados: Treino e Teste

Para se ter um modelo genrico, isto , que atenda dados do mundo real da melhor maneira possvel,  necessrio realizar a diviso do conjunto de dados em grupo de dados de treino, com os quais o modelo ir aprender, e, outro conjunto denominado de testes, que servir para avaliarmos o desempenho do modelo criado.

Essa diviso deve ocorrer de forma aleatria para evitar amostras enviesadas. Bem como, ela tem que ser feita antes do balanceamento dos dados, para que se possa confirmar que o balanceamento est adequado.

Antes dessa etapa,  necessrio segregar os dados da classe alvo, isto , da varivel `Class` que  a que iremos prever. Isso serve para que os dados no contenham informaes que o auxiliem a identificar os tipos de transaes. Para isso separa-se as variveis independentes em *X*, e a varivel alvo (`Class`) em uma varivel *y*.

A diviso dos dados em Treino e Teste contm algumas configuraes adicionais:
- o tamanho foi definido em 70:30, o que significa que o conjunto de Treino conter 70% do total dos dados, enquanto que o conjunto de Testes ficar com 30% do total do conjunto de  dados;
- os conjuntos de Treino e Teste contero a mesma quantidade de classes,  proporcionalmente, ou seja, a mesma quantidade de transaes legtimas e transaes fraudulentas;
- a randomizao dos dados foi ativada, para misturar os registros e, assim, garantir dados aleatrios nos conjuntos de Treino e de Teste;
- e, foi indicado um valor de *seed* que ir proporcionar a reproduo do cdigo sem alteraes no resultado.

```python
# separar variveis independentes da varivel dependente
X = df_clean.drop('Class', axis=1)
y = df['Class']

# dividir dados de treino e de teste
## stratify= y (para dividir de forma que as classes tenham a mesma proporo)
## random_state para que o resultado seja replicvel
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3,
                                                    stratify=y, shuffle=True,
                                                    random_state=110)
```

### Balanceamento dos Dados

Como visto, o conjunto de dados est desbalanceado e  necessrio re-balancear os dados para evitar a criao de um modelo que tenha baixa performance na identificao de transaes fraudulentas, bem como o *overfitting* e, assim, criarmos um bom modelo de *machine learning*.
<br/>
<br/>

O mtodo utilizado para ser aplicado nesse caso  o ***Undersampling***, que reduz a classe majoritria, excluindo esses dados de forma aleatria. Dessa forma, as caractersticas da classe minoritria, que nesse caso so os dados de transaes fraudulentas, so preservados. E so esses os dados mais importantes para a soluo do nosso problema.
<br/>
<br/>

Essa tcnica  aplicada apenas no conjunto de treino para que as caractersticas do conjunto de testes no sejam desconfiguradas.

```python
# usar tcnica undersampling
## random_state para que o resultado seja replicvel
rus = RandomUnderSampler(random_state=606)
X_rus, y_rus = rus.fit_resample(X_train, y_train)

# ver o balanceamento das classes
print(pd.Series(y_rus).value_counts())
```

Entre essas mtricas, o Recall  a que nos fornece a melhor medida para o problema especfico em estudo. Em outras palavras,  melhor que o modelo erre ao dizer que uma transao  fraude, quando na realidade ela no seja. Do que erre ao apontar uma transao como legtima, quando na verdade  fraudulenta, trazendo prejuzos para os negcios.

$${\text{Recall: }}\frac{\text{Verdadeiro Positivo}} {\text{Verdadeiro Positivo+ Falso Negativo}}$$

## Regresso Logstica

A Regresso Logstica  um algoritmo de classificao que, conforme feita as observaes, ir atribu-las a uma classe de acordo com a sua probabilidade de estar nela e dentro de um grupo de classes. Por isso, faz dele um bom modelo para ser utilizado quando a varivel dependente  categrica.

No nosso estudo, isso significa que o modelo, uma vez treinado, ir receber a informao de uma nova transao de carto de crdito e, de acordo com o algoritmo, vai determinar qual a porcentagem dela ser ou no uma fraude. Com isso, a maior probabilidade de classe  o que o algoritmo ir apontar como a possvel classe dessa transao.

Para fazer isso, a regresso logstica transforma a sada gerada pelo modelo, usando a funo sigmoide (que  uma funo logstica), para retornar um valor em probabilidade e ento, definir a qual classe pertence a observao.

### Treinamento do Modelo

## rvore de Deciso

Do ingls, *Decision Tree*, ela  um algoritmo de aprendizado supervisionado utilizado para problemas de classificaes e de regresses. Seu processo envolve encontrar limites nos dados e, ento, dividi-los em subconjuntos.

### Treinamento do Modelo

Para criar um modelo de rvore de Deciso  necessrio informar um hiperparmetro que informa a profundidade mxima que a rvore deve ir. Como no existe uma mtrica para decidir isso, apenas uma recomendao de que se inicie com 3 e depois sejam testados valores diferentes, vou fazer isso inicialmente. Isto , criar diversos modelos com profundidade diferentes e coletar os dados de Recall de cada modelo. Com isso consigo saber qual a profundidade vai me gerar o modelo com melhor Recall.

A deciso de escolher o modelo com o melhor Recall se deve ao fato de que  uma das principais mtricas para sem observadas no caso do nosso estudo. Uma vez que, como visto, errar em apontar um caso de fraude  extremamente prejudicial para os negcios. Portanto buscamos uma baixa taxa de Falsos Negativos, consequentemente, queremos uma alta taxa de Recall.

```python
# criar modelos com rvore de Deciso

## criar lista para salvar resultados do Recall
depth_final = []

## loop para gerar modelos com diferentes profundidades
for i in range(3, 11):
  # selecionar seed para replicar o modelo
  np.random.seed(29)

  # instanciar e configurar hiperparmetros
  ## profundidade = de 3 a 10
  ## critrio para definir as folhas = entropia
  model_DT = DecisionTreeClassifier(max_depth=i, criterion='entropy')

  # treinar o modelo nos dados de treino
  model_DT.fit(X_rus, y_rus)

  # fazer previses
  y_pred_DT = model_DT.predict(X_test)

  # calcular Recall e adicion-lo  lista
  depth_final.append(recall_score(y_test, y_pred_DT))

# imprimir melhor Recall
print('O maior valor de Recall encontrado foi de {:.4f}.'.format(max(depth_final)))
print('Posio na lista: \t {}'.format(depth_final.index(max(depth_final))))
print('Profundidade: \t\t {}'.format((depth_final.index(max(depth_final))) + 3))
```

Agora que encontramos a melhor profundidade para  maior taxa de Recall, podemos rodar o modelo:

```python
# criar modelo de rvore de Deciso com melhor profundidade

# selecionar seed para replicar o modelo
np.random.seed(29)

# instanciar e configurar hiperparmetros
## nvel de profundidade = 6
## critrio para definir as folhas = entropia
model_DT_final = DecisionTreeClassifier(max_depth=6, criterion='entropy')

# treinar o modelo nos dados de treino
model_DT_final.fit(X_rus, y_rus)

# fazer previses
y_pred_DT_final = model_DT_final.predict(X_test)
```

```python
# calcular a probabilidade das classes
y_prob_DT_final = model_DT_final.predict_proba(X_test)

# imprimir as probabilidades dos 5 primeiros registros
print('y_pred_DT_final \n', y_pred_DT_final[0:5])
print('\n y_prob_DT_final \n', y_prob_DT_final[0:5])
```

Acima, podemos ver que o resultado gerado pela rvore de Deciso  semelhante ao da Regresso Logstica. Ou seja,  gerada uma lista com os resultados para cada transao onde  dada a probabilidade dela ser legtima, ou de ser fraude. Na primeira linha o resultado foi de [1. 0.] isso significa que a probabilidade de ser uma transao legtima  de 100%, enquanto que a chance de ser fraude  0%.
