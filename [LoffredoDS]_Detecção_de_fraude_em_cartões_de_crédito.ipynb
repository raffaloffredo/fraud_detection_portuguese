# Obtenção dos Dados

Os dados utilizados nesse projeto foram disponibilizados por operadoras europeias de cartão de crédito. Esse conjunto de dados contém transações referentes à um período de 2 dias, que resultaram em quase 290 mil transações, nas quais 492 foram classificadas como fraudes. Ou seja, trata-se de um conjunto de dados desbalanceado uma vez que as fraudes representam cerca de 0,17% do total de transações.
<br/>
<br/>

Na [página original dos dados](https://www.kaggle.com/datasets/mlg-ulb/creditcardfraud) é informado ainda que todos os atributos foram convertidos para dados numéricos por questões de proteção de dados e descaracterização dos clientes. . Mesmo o nome das variáveis foi modificado e é identificado apenas pela sequência da letra 'V' seguida do número 1, 2, 3, ..., 28. Ou seja, o que se vê é apenas V1, V2, V3, ..., V28.
<br/>
<br/>

Adiciona-se ainda, o fato de que essas variáveis passaram pelo processo de transformação com o algoritmo não-supervisionado denominado PCA - *Principal Component Analysis* - que tem por objetivo reduzir a dimensionalidade. Isto é, um processo que visa diminuir a quantidade original de atributos no conjunto de dados com a menor perda de informações possível. Além disso, o PCA consegue distinguir as informações relevantes daqueles aleatórias e, até mesmo, redundantes. Por tomar conhecimento desse fato, conclui-se que os dados estão padronizados (estão todos em uma mesma escala), uma vez que isso é pré-requisito para se aplicar o PCA.
<br/>
<br/>

Por fim, o site ressalta que as variáveis `Time` e `Amount` não passaram pelo processo de transformação mencionado acima. O atributo `Time` refere-se ao tempo decorrido entre a cada transação e a primeira transação do conjunto de dados. Já a característica `Amount` é o valor monetário de cada transação. Temos também a classe desse conjunto, isso é, o nosso alvo, que responde a pergunta do nosso problema: "Essa transação é uma fraude?". Para essa variável a resposta de "Sim" ou "Não" está convertida em um valor binário de 0 quando não se trata de fraude e 1 para quando for reconhecida como fraude.

Ou seja, o modelo prevê que uma transação não é fraudulenta, quando na verdade ela é! Por termos essas 3 variáveis conhecidas, vamos estudá-las um pouco mais a fundo, a fim de tentar identificar padrões nas transações fraudulentas. Dessa forma, vamos checar a distribuição de frequência das transações fraudulentas e das legítimas ao longo do tempo de dados que temos disponíveis.


Primeiro, vou verificar quanto tempo durou o recolhimento dos dados.

```python
# verificar tempo de duração da coleta de dados
print('Duração total da coleta dos dados: {:.0f} segundos'.format(df.Time.max()))
print('Isso corresponde a {:.0f} horas.'.format(df.Time.max() / 3600))
```

Portanto, temos 48h de coleta de dados. Poderemos checar isso hora-a-hora.

```python
# plotar histograma de transações legítimas x fraudulentas ao longo do tempo

## definir variável com transações legítimas x fraudes
legitimate = df[df.Class == 0].Time.div(3600)
fraud = df[df.Class == 1].Time.div(3600)

## configurar histograma
fig, ax = plt.subplots(nrows=2, ncols=1, figsize=(10,4))

## título
ax[0].text(-1.5, 14000, 'Frequência de Transações', fontsize=20, color='#004a8f',
        fontweight='bold')

## subtítulo
ax[0].text(-1.5, 12500, 'Transações realizadas com cartão de crédito por hora',
           fontsize=9, color='#6d6e70')

## dados
### gráfico 1 - transações legítimas
ax[0].hist(legitimate, label='Legítimas', bins=48, color='#bdbdbd')
ax[0].legend(loc='upper left', frameon=False)
ax[0].spines['right'].set_visible(False)
ax[0].spines['top'].set_visible(False)
ax[0].spines['bottom'].set_visible(False)
ax[0].spines['left'].set_visible(False)
ax[0].set_yticklabels(labels=[], visible=False)
ax[0].set_yticks(ticks=[])

### gráfico 2 - transacões fraudulentas
ax[1].hist(fraud, label='Fraudes', bins=48, color='#004a8f')
ax[1].legend(loc='upper left', frameon=False)
ax[1].spines['right'].set_visible(False)
ax[1].spines['top'].set_visible(False)
ax[1].spines['bottom'].set_visible(False)
ax[1].spines['left'].set_visible(False)
ax[1].set_yticklabels(labels=[], visible=False)
ax[1].set_yticks(ticks=[])

plt.show()
```

Observa-se que as transações legítimas ocorrem com mais fluidez e em determinados horários. Como não sabemos a hora exata que se iniciaram a coleta dos dados, apenas podemos levantar uma hipótese de que seriam horários comerciais, onde a população de forma geral está mais ativa.


Por outro lado, no histograma com a frequência de fraudes, temos o oposto: não há uma fluidez pois o gráfico tem muitos picos. Também não há conglomerados de horários em que se tem mais transações fraudulentas e, tampouco horários com menos dessas transações. Logo, por não apresentar um padrão, fica mais difícil reconhecê-las.

Vamos agora checar se é possível identificar um padrão nas transações fraudulentas, com base na quantidade monetária a ser paga pelo cartão de crédito. Para isso, com a variável `Amount`, vou plotar um gráfico de frequências (histograma) e também um gráfico do tipo *boxplot*, uma vez que eles trazem mais informações estatísticas para analisar. # Tratamento dos Dados

Antes de prosseguir para a criação dos modelos de *machine learning* é necessária uma etapa preliminar de preparação dos dados para que possam gerar os melhores algoritmos. São 3 operações que precisaremos nos atentar uma vez que a ordem deles importa:
1. padronizar os atributos `Time` e `Amount`;
2. dividir os dados em dois conjuntos: um de treino e outro de testes;
3. balancear o conjunto de dados.



E antes de realizar alterações no conjunto de dados, vou copiá-lo, para que as alterações feita daqui por diante sejam realizadas nessa réplica. Assim, o conjunto original é mantido intacto.

```python
# fazer cópia do data set
df_clean = df.copy()
```

### Padronizar Variáveis

A padronização dos dados tem como objetivo deixar todos os valores em uma mesma escala para que atributos diferentes possam ser comparados entre si. Quando os dados não estão padronizados algumas variáveis podem causar mais impacto do que outras, prejudicando a performance do modelo já que ele irá dar vantagem a uma variável em prol de outras. Logo, a padronização evita que o modelo fique enviesado.

Como visto, as variáveis `V1, V2, V3, ..., V28` já passaram pelo PCA e, portanto, infere-se que foram padronizadas, uma vez que esse processo é requisito para a aplicação do PCA nos dados. Portanto, restam as variáveis `Time` e `Amount` que não estão padronizadas.

Essa etapa, em condições normais, seria feita em dois momentos diferentes, já com os dados divididos em treino e teste (nos dados de treino, antes de treinar o modelo e, nos dados de teste, antes de serem feitas as previsões). Contudo, como o conjunto de dados já passou por um tratamento semelhante em outros dados, portanto, realizar essa etapa antes, não irá trazer maiores prejuízos.

Dessa forma, optou-se por utilizar o ***StandardScaler*** uma vez que ele segue a distribuição normal e transforma a média em 0 e o desvio padrão em 1.

```python
# padronizar atributos 'Time' e 'Amount' com StandarScaler
## instanciar StandardScaler
std_scaler = StandardScaler()

# criar nova coluna com os dados padronizados em 'Time'
df_clean['Time_STD'] = std_scaler.fit_transform(df_clean['Time'].values.reshape(-1, 1))

# criar nova coluna com os dados padronizados em 'Amount'
df_clean['Amount_STD'] = std_scaler.fit_transform(df_clean['Amount'].values.reshape(-1, 1))

# excluir colunas 'Time' e 'Amount' sem padronização
df_clean.drop(['Time', 'Amount'], axis=1, inplace=True)

# verificar primeiras entradas do data set modificado
df_clean.head()
```

Podemos checar o resultado da padronização nos atributos `Time_STD` e `Amount_STD`, com o resumo estatístico:

```python
# imprimir resumo estatístico para 'Time_STD' e 'Amount_STD'
df_clean[['Time_STD', 'Amount_STD']].describe().round(3)
```

Em ambos veja que o Standard Scaler cumpriu sua função e, agora, temos uma média igual a 0 e um desvio padrão igual a 1.

### Divisão dos Dados: Treino e Teste

Para se ter um modelo genérico, isto é, que atenda dados do mundo real da melhor maneira possível, é necessário realizar a divisão do conjunto de dados em grupo de dados de treino, com os quais o modelo irá aprender, e, outro conjunto denominado de testes, que servirá para avaliarmos o desempenho do modelo criado.

Essa divisão deve ocorrer de forma aleatória para evitar amostras enviesadas. Bem como, ela tem que ser feita antes do balanceamento dos dados, para que se possa confirmar que o balanceamento está adequado.

Antes dessa etapa, é necessário segregar os dados da classe alvo, isto é, da variável `Class` que é a que iremos prever. Isso serve para que os dados não contenham informações que o auxiliem a identificar os tipos de transações. Para isso separa-se as variáveis independentes em *X*, e a variável alvo (`Class`) em uma variável *y*.

A divisão dos dados em Treino e Teste contém algumas configurações adicionais:
- o tamanho foi definido em 70:30, o que significa que o conjunto de Treino conterá 70% do total dos dados, enquanto que o conjunto de Testes ficará com 30% do total do conjunto de  dados;
- os conjuntos de Treino e Teste conterão a mesma quantidade de classes,  proporcionalmente, ou seja, a mesma quantidade de transações legítimas e transações fraudulentas;
- a randomização dos dados foi ativada, para misturar os registros e, assim, garantir dados aleatórios nos conjuntos de Treino e de Teste;
- e, foi indicado um valor de *seed* que irá proporcionar a reprodução do código sem alterações no resultado.

```python
# separar variáveis independentes da variável dependente
X = df_clean.drop('Class', axis=1)
y = df['Class']

# dividir dados de treino e de teste
## stratify= y (para dividir de forma que as classes tenham a mesma proporção)
## random_state para que o resultado seja replicável
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3,
                                                    stratify=y, shuffle=True,
                                                    random_state=110)
```

### Balanceamento dos Dados

Como visto, o conjunto de dados está desbalanceado e é necessário re-balancear os dados para evitar a criação de um modelo que tenha baixa performance na identificação de transações fraudulentas, bem como o *overfitting* e, assim, criarmos um bom modelo de *machine learning*.
<br/>
<br/>

O método utilizado para ser aplicado nesse caso é o ***Undersampling***, que reduz a classe majoritária, excluindo esses dados de forma aleatória. Dessa forma, as características da classe minoritária, que nesse caso são os dados de transações fraudulentas, são preservados. E são esses os dados mais importantes para a solução do nosso problema.
<br/>
<br/>

Essa técnica é aplicada apenas no conjunto de treino para que as características do conjunto de testes não sejam desconfiguradas.

```python
# usar técnica undersampling
## random_state para que o resultado seja replicável
rus = RandomUnderSampler(random_state=606)
X_rus, y_rus = rus.fit_resample(X_train, y_train)

# ver o balanceamento das classes
print(pd.Series(y_rus).value_counts())
```

```python
# plotar gráfico de barras para transações legítimas x fraudes
## definir eixos
x = ['Legítimas', 'Fraudes']
y = y_rus.value_counts()

## configurar cores das barras
bar_colors = ['#bdbdbd', '#004a8f']

## plotar gráfico
fig, ax = plt.subplots(figsize=(6, 5))
ax.bar(x, y, color=bar_colors)
ax.set_yticklabels(labels=[], visible=False)

### título
ax.text(-0.5, 410, 'Resultado do Balanceamento', fontsize=20, color='#004a8f',
        fontweight='bold')

### subtítulo
ax.text(-0.5, 370, 'Quantidade de registros em cada classe alvo, após o'
        '\nbalanceamento com o Standard Scale\n', fontsize=8, color='#6d6e70')

### quantidade de transações legítimas
ax.text(-0.1, 325, '344', fontsize=12, color="#6d6e70", fontweight='bold')

### quantidade de transações fraudulentas
ax.text(0.9, 325, '344', fontsize=12, color="#FFFFFF", fontweight='bold')

plt.show()
```

Após a aplicação do método *Undersampling*, vemos que ambas as classes agora tem a mesma proporção. Isto é, para transações legítimas temos 344 registros e, para as fraudes, o mesmo valor.

## Métricas de Avaliação

Para avaliar o desempenho de algoritmos de classificação as métricas utilizadas são Precisão, Recall, acurácia, F1-Score e a matriz de confusão. Contudo, quando se tem dados desbalanceados, como é o caso do conjunto em estudo, ainda que ele tenha sido já tratado e balanceado, a acurácia não é uma boa métrica de avaliação. Isso se deve ao fato de que se pode obter um resultado de acurácia muito alto, entretanto, na identificação de fraudes a detecção ser muito baixa, o que não nos leva ao nosso objetivo.
<br/>
<br/>

Entre essas métricas, o Recall é a que nos fornece a melhor medida para o problema específico em estudo. Isso se deve porque, no caso de fraude, os Falsos Negativos são mais prejudiciais para uma empresa do que os Falsos Positivos. Em outras palavras, é melhor que o modelo erre ao dizer que uma transação é fraude, quando na realidade ela não seja. Do que erre ao apontar uma transação como legítima, quando na verdade é fraudulenta, trazendo prejuízos para os negócios.

Tendo isso em vista, busca-se uma alta taxa de Recall.

O Recall olha para todas as transações legítimas e visa responder: o quanto o modelo acerta? O resultado será um valor entre 0 e 1, sendo que, quanto mais próximo de 1, melhor, pois indica uma baixa taxa de Falsos Negativos.
<br/>
<br/>

Seu cálculo é dado por:
<br/>
<br/>

$${\text{Recall: }}\frac{\text{Verdadeiro Positivo}} {\text{Verdadeiro Positivo+ Falso Negativo}}$$
<br/>
<br/>

Ainda pensando na finalidade desse estudo, outra métrica de avaliação de modelos de classificação que pode ser utilizada é a AUC - *Area Under the Curve*, traduzindo é a Area abaixo da curva. Ela deriva do ROC - *Receiver Operating Characteristic* - e indica para nós o quanto o modelo consegue distinguir entre duas coisas. No nosso exemplo, entre uma transação legítima de uma fraude. O valor dado por AUC varia entre 0 e 1, sendo que, quanto mais próximo de 1, indica que melhor é o modelo.

<br/>

## Regressão Logística

A Regressão Logística é um algoritmo de classificação que, conforme feita as observações, irá atribuí-las a uma classe de acordo com a sua probabilidade de estar nela e dentro de um grupo de classes. Por isso, faz dele um bom modelo para ser utilizado quando a variável dependente é categórica.

No nosso estudo, isso significa que o modelo, uma vez treinado, irá receber a informação de uma nova transação de cartão de crédito e, de acordo com o algoritmo, vai determinar qual a porcentagem dela ser ou não uma fraude. Com isso, a maior probabilidade de classe é o que o algoritmo irá apontar como a possível classe dessa transação.

Para fazer isso, a regressão logística transforma a saída gerada pelo modelo, usando a função sigmoide (que é uma função logística), para retornar um valor em probabilidade e então, definir a qual classe pertence a observação.

## Árvore de Decisão

Do inglês, *Decision Tree*, ela é um algoritmo de aprendizado supervisionado utilizado para problemas de classificações e de regressões. Seu processo envolve encontrar limites nos dados e, então, dividi-los em subconjuntos.

### Treinamento do Modelo

Para criar um modelo de Árvore de Decisão é necessário informar um hiperparâmetro que informa a profundidade máxima que a árvore deve ir. Como não existe uma métrica para decidir isso, apenas uma recomendação de que se inicie com 3 e depois sejam testados valores diferentes, vou fazer isso inicialmente. Isto é, criar diversos modelos com profundidade diferentes e coletar os dados de Recall de cada modelo. Com isso consigo saber qual a profundidade vai me gerar o modelo com melhor Recall.

A decisão de escolher o modelo com o melhor Recall se deve ao fato de que é uma das principais métricas para sem observadas no caso do nosso estudo. Uma vez que, como visto, errar em apontar um caso de fraude é extremamente prejudicial para os negócios. Portanto buscamos uma baixa taxa de Falsos Negativos, consequentemente, queremos uma alta taxa de Recall.

```python
# criar modelos com Árvore de Decisão

## criar lista para salvar resultados do Recall
depth_final = []

## loop para gerar modelos com diferentes profundidades
for i in range(3, 11):
  # selecionar seed para replicar o modelo
  np.random.seed(29)

  # instanciar e configurar hiperparâmetros
  ## profundidade = de 3 a 10
  ## critério para definir as folhas = entropia
  model_DT = DecisionTreeClassifier(max_depth=i, criterion='entropy')

  # treinar o modelo nos dados de treino
  model_DT.fit(X_rus, y_rus)

  # fazer previsões
  y_pred_DT = model_DT.predict(X_test)

  # calcular Recall e adicioná-lo à lista
  depth_final.append(recall_score(y_test, y_pred_DT))

# imprimir melhor Recall
print('O maior valor de Recall encontrado foi de {:.4f}.'.format(max(depth_final)))
print('Posição na lista: \t {}'.format(depth_final.index(max(depth_final))))
print('Profundidade: \t\t {}'.format((depth_final.index(max(depth_final))) + 3))
```

Agora que encontramos a melhor profundidade para  maior taxa de Recall, podemos rodar o modelo:

```python
# criar modelo de Árvore de Decisão com melhor profundidade

# selecionar seed para replicar o modelo
np.random.seed(29)

# instanciar e configurar hiperparâmetros
## nível de profundidade = 6
## critério para definir as folhas = entropia
model_DT_final = DecisionTreeClassifier(max_depth=6, criterion='entropy')

# treinar o modelo nos dados de treino
model_DT_final.fit(X_rus, y_rus)

# fazer previsões
y_pred_DT_final = model_DT_final.predict(X_test)
```

```python
# calcular a probabilidade das classes
y_prob_DT_final = model_DT_final.predict_proba(X_test)

# imprimir as probabilidades dos 5 primeiros registros
print('y_pred_DT_final \n', y_pred_DT_final[0:5])
print('\n y_prob_DT_final \n', y_prob_DT_final[0:5])
```

Acima, podemos ver que o resultado gerado pela Árvore de Decisão é semelhante ao da Regressão Logística. Ou seja, é gerada uma lista com os resultados para cada transação onde é dada a probabilidade dela ser legítima, ou de ser fraude. Na primeira linha o resultado foi de [1. 0.] isso significa que a probabilidade de ser uma transação legítima é de 100%, enquanto que a chance de ser fraude é 0%.
